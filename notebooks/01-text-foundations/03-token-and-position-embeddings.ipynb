# 03 - Token and Position Embeddings

Map token IDs to learned vectors, map sequence positions to learned vectors, and add the two representations.

```python
import torch

torch.manual_seed(123)
token_ids = torch.tensor([[2, 3, 5, 1], [3, 1, 0, 2]])
vocab_size = 6
embedding_dim = 3
context_length = token_ids.shape[1]

token_embedding = torch.nn.Embedding(vocab_size, embedding_dim)
position_embedding = torch.nn.Embedding(context_length, embedding_dim)

token_vectors = token_embedding(token_ids)
position_vectors = position_embedding(torch.arange(context_length))
input_vectors = token_vectors + position_vectors

print('token vectors:', token_vectors.shape)
print('position vectors:', position_vectors.shape)
print('combined inputs:', input_vectors.shape)
assert input_vectors.shape == (2, 4, 3)
```

The position tensor has shape `(context_length, embedding_dim)`. PyTorch broadcasts it across the batch dimension when it is added to the token tensor.
